```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("/content/diabetes.csv")
x=df.iloc[::-1]
y=df['Outcome']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=2/3,random_state=2)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report,roc_auc_score,roc_curve
cn_mtx=confusion_matrix(y_test,y_pred)
a=ConfusionMatrixDisplay(confusion_matrix=cn_mtx)
a.plot()
print("accuracy ;",accuracy_score(y_test,y_pred))
print("Classification report ;",classification_report(y_test,y_pred))
```
